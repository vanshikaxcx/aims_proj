```python
%%writefile owl_inference.py
import os
import torch
import streamlit as st
from PIL import Image, ImageDraw
from transformers import OwlViTProcessor, OwlViTForObjectDetection

# --- Config ---
MODEL_DIR = "weights/owl_model_epoch_20.pth"  # Fine-tuned weights
MODEL_NAME = "google/owlvit-base-patch32"     # Base model
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
CONF_THRESHOLD = 0.5

@st.cache_resource
def load_model():
    processor = OwlViTProcessor.from_pretrained(MODEL_NAME)
    model = OwlViTForObjectDetection.from_pretrained(MODEL_NAME)
    model.load_state_dict(torch.load(MODEL_DIR, map_location=DEVICE))
    model = model.to(DEVICE).eval()
    return processor, model

def run_inference(processor, model, image, queries):
    inputs = processor(text=queries, images=image, return_tensors="pt").to(DEVICE)
    with torch.no_grad():
        outputs = model(**inputs)
    target_sizes = torch.tensor([image.size[::-1]])  # (height, width)
    results = processor.post_process_object_detection(outputs, target_sizes=target_sizes)

    detections = []
    for score, label, box in zip(results[0]["scores"], results[0]["labels"], results[0]["boxes"]):
        score = score.item()
        if score >= CONF_THRESHOLD:
            box = [round(coord, 2) for coord in box.tolist()]
            detections.append({
                "label": queries[label],
                "score": round(score, 3),
                "box": box
            })
    return detections

def draw_boxes(image, detections):
    draw = ImageDraw.Draw(image)
    for det in detections:
        box = det["box"]
        label = f"{det['label']} ({det['score']*100:.1f}%)"
        draw.rectangle(box, outline="red", width=3)
        draw.text((box[0], box[1] - 10), label, fill="red")
    return image

# --- Streamlit UI ---
st.title("OWL-ViT Object Detection")
st.write("Upload an image and enter your queries to detect objects using your fine-tuned model.")

processor, model = load_model()

uploaded_file = st.file_uploader("Upload an image", type=["jpg", "png", "jpeg"])
queries_input = st.text_input("Enter object queries (comma separated)", "person, dog, car")

if uploaded_file and queries_input:
    image = Image.open(uploaded_file).convert("RGB")
    queries = [q.strip() for q in queries_input.split(",")]

    st.image(image, caption="Uploaded Image", use_container_width=True)

    with st.spinner("Running detection..."):
        detections = run_inference(processor, model, image, queries)

    if detections:
        result_img = draw_boxes(image.copy(), detections)
        st.image(result_img, caption="Detection Results", use_container_width=True)
        st.write("Detections:", detections)
    else:
        st.warning("No objects detected above the confidence threshold.")
```

```python
# Step 1: Create the owl_inference.py file
%%writefile owl_inference.py

import os
import torch
from PIL import Image, ImageDraw
from transformers import OwlViTProcessor, OwlViTForObjectDetection
import numpy as np

# --- Configuration ---
MODEL_DIR = "weights/owl_model_epoch_20.pth"
MODEL_NAME = "google/owlvit-base-patch32"
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
CONF_THRESHOLD = 0.1

# --- Load Models (once) ---
def load_owlvit_model():
    """Loads the fine-tuned OWL-ViT model and processor."""
    print("Loading fine-tuned OWL-ViT model...")
    processor = OwlViTProcessor.from_pretrained(MODEL_NAME)
    model = OwlViTForObjectDetection.from_pretrained(MODEL_NAME)
    if os.path.exists(MODEL_DIR):
        model.load_state_dict(torch.load(MODEL_DIR, map_location=DEVICE))
        print(f"Loaded fine-tuned weights from {MODEL_DIR}")
    else:
        print("Fine-tuned weights not found. Using pre-trained model.")

    model.to(DEVICE).eval()
    return processor, model

# --- Inference Logic ---
def get_initial_prediction(processor, model, image, query):
    """Runs OWL-ViT inference to get the best coarse bounding box."""
    inputs = processor(text=query, images=image, return_tensors="pt").to(DEVICE)
    with torch.no_grad():
        outputs = model(**inputs)

    target_sizes = torch.tensor([image.size[::-1]])
    results = processor.post_process_object_detection(outputs, target_sizes=target_sizes, threshold=CONF_THRESHOLD)

    if not results[0]['scores'].numel():
        return None, None

    best_score_idx = results[0]['scores'].argmax()
    best_box = results[0]['boxes'][best_score_idx].cpu().numpy()

    return best_box.tolist(), results[0]['scores'][best_score_idx].item()

def draw_boxes(image, detections):
    """Draws bounding boxes on an image and returns the annotated image."""
    draw = ImageDraw.Draw(image)
    if detections:
        for det in detections:
            box = det['box']
            label = f"{det.get('label', 'object')} ({det['score'] * 100:.1f}%)"
            draw.rectangle(box, outline=det.get('color', 'red'), width=3)
            draw.text((box[0], box[1] - 10), label, fill=det.get('color', 'red'))
    return image
```

```python
# Step 2: Create the app.py file
%%writefile app.py

import os
import torch
from PIL import Image, ImageDraw
import numpy as np
import ipywidgets as widgets
from IPython.display import display, clear_output
import io
from owl_inference import load_owlvit_model, get_initial_prediction, draw_boxes

# We need to import the SAM model components.
# Assuming you have sam_vit_h_4b8939.pth in your environment.
from segment_anything import sam_model_registry, SamPredictor

# --- SAM Configuration ---
SAM_CHECKPOINT = "sam_vit_h_4b8939.pth"
SAM_MODEL_TYPE = "vit_h"
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# --- Load Models (once) ---
print("Loading OWL-ViT model...")
owlvit_processor, owlvit_model = load_owlvit_model()

print("Loading SAM model...")
sam = sam_model_registry[SAM_MODEL_TYPE](checkpoint=SAM_CHECKPOINT).to(device=DEVICE)
sam_predictor = SamPredictor(sam)
print("SAM loaded.")

# --- UI Widgets ---
uploader = widgets.FileUpload(accept='image/*', description='Upload Image')
text_prompt_input = widgets.Text(value='a person', description='Prompt:')
run_button = widgets.Button(description='Run Pipeline', button_style='success')
reset_button = widgets.Button(description='Reset', button_style='warning')
output_area = widgets.Output()

# --- Full Pipeline Function ---
def run_full_pipeline(b):
    output_area.clear_output()

    if not uploader.value:
        with output_area:
            print("Please upload an image first.")
        return

    uploaded_file_info = next(iter(uploader.value.values()))
    image_source = Image.open(io.BytesIO(uploaded_file_info['content'])).convert("RGB")

    text_prompt = text_prompt_input.value
    if not text_prompt:
        with output_area:
            print("Please enter a text prompt.")
        return

    # --- Stage 1: OWL-ViT for Coarse Localization ---
    initial_box, score = get_initial_prediction(owlvit_processor, owlvit_model, image_source, [text_prompt])

    if initial_box is None:
        with output_area:
            print(f"No objects detected for prompt '{text_prompt}' above threshold.")
            display(image_source)
        return

    # --- Stage 2: SAM for Fine-grained Refinement ---
    image_np = np.array(image_source)
    sam_predictor.set_image(image_np)
    masks, scores, logits = sam_predictor.predict(box=np.array(initial_box), multimask_output=True)
    best_mask = masks[np.argmax(scores)]

    # --- Stage 3: Recalculate Bounding Box from Mask ---
    y_indices, x_indices = np.where(best_mask)
    if len(x_indices) == 0:
        refined_box = initial_box
    else:
        x_min, x_max = np.min(x_indices), np.max(x_indices)
        y_min, y_max = np.min(y_indices), np.max(y_indices)
        refined_box = [x_min, y_min, x_max, y_max]

    # --- Visualize Results ---
    result_image = image_source.copy()

    draw_initial = ImageDraw.Draw(result_image)
    draw_initial.rectangle(initial_box, outline="yellow", width=3)

    draw_refined = ImageDraw.Draw(result_image)
    draw_refined.rectangle(refined_box, outline="red", width=3)

    with output_area:
        print(f"Result for prompt: '{text_prompt}'")
        print("Yellow Box: Initial OWL-ViT Prediction | Red Box: SAM Refined Prediction")
        display(result_image)

def reset_ui(b):
    uploader.value.clear()
    output_area.clear_output()
    text_prompt_input.value = 'a person'
    with output_area:
        print("UI Reset. Please upload a new image.")

# Link buttons to functions
run_button.on_click(run_full_pipeline)
reset_button.on_click(reset_ui)

# --- Display the UI ---
def display_ui():
    buttons = widgets.HBox([run_button, reset_button])
    display(widgets.VBox([uploader, text_prompt_input, buttons, output_area]))

display_ui()
```
